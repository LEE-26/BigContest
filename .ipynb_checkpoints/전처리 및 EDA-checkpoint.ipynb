#### target 과 crosstab

```python
def crosstab_apply(df, column1 , column2, criterion = None) :
    if criterion == "index" :
        print(column1, "기준입니다.")
        table = np.round(pd.crosstab(index = df[column1], columns = df[column2], normalize = criterion) * 100, 2)
    elif criterion == "columns": 
        print(column2, "기준입니다.")
        table = np.round(pd.crosstab(index = df[column1], columns = df[column2], normalize = criterion) * 100, 2)
        
    else : 
        table = pd.crosstab(index = df[column1], columns = df[column2])
    
    return table
    
```

### % birth_year => age_group (파생변수)

```python
# age
train_dset["age"] = train_dset["birth_year"].apply(lambda x : 2022 - x + 1) 
test_dset["age"] = test_dset["birth_year"].apply(lambda x : 2022 - x + 1) 
minimum_age = train_dset["age"].min()
maximum_age = train_dset["age"].max()

minimum_age_test = test_dset["age"].min()
maximum_age_test = test_dset["age"].max()

print("최소 나이 : {}살".format(minimum_age))
print("최대 나이 : {}살".format(maximum_age))

print("최소 나이 test_dset: {}살".format(minimum_age_test))
print("최대 나이 test_dset: {}살".format(maximum_age_test))


# 연령대 categorical 변수 생성

def age_group(x) : 
    if 20 <= x < 30 : return 2
    elif 30 <= x < 40 : return 3
    elif 40 <= x < 50 : return 4
    elif 50 <= x < 60 : return 5
    elif 60 <= x < 70 : return 6 
    elif 70 <= x < 80 : return 7
    elif 80 <= x < 90 : return 8
    else : return 9         
    
train_dset["age_group"] = train_dset["age"].apply(lambda x : age_group(x))
test_dset["age_group"] = test_dset["age"].apply(lambda x : age_group(x))
train_dset["age_group"].value_counts()
```

## Purpose(대출목적)
- 대환대출 : 금융기관에서 대출을 받아 이전의 대출금이나 연체금을 갚는 제도를 말한다. 특히 신용불량자 또는 신용카드 대금 연체자가 밀린 연체대금을 장기대출로 바꾸어 분할 납부하도록 하는 경우를 말한다. 이럴 경우 연체자는 신용불량자로 등록되는 것을 미리 막을 수 있는 도움을 주지만 금융기관으로부터 대환대출을 받기 위해서는 그동안의 대출 실적이나 신용도가 우수해야할 뿐만 아니라 대환대출의 이자가 보통 은행 대출 금리보다 높다는 점을 감수해야 한다. 반면 신용카드사들은 연체율 또는 부실채권을 줄일 수 있게 된다. A 금융기관에서 대출을 받아 B 금융기관의 부채를 갚는 대환대출 금융거래는 저금리대환대출, 아파트대환대출과 같은 변형된 대환대출도 만들어 냈다. 금리 변동에 따른 금융비용의 부담이나 아파트를 담보로 대출 받았지만 타 금융사의 담보대출 조건이 금리나 상환 조건 면에서 더 유리할 경우 대환대출을 받는 경우를 말한다.

## loan_limit 
- (히스토그램 및 연속형 변수 시각화)

```python
# 이상치를 제거하고 histogram 그려보기
Q3 = train_dset["loan_limit"].quantile(0.75)
Q1 = train_dset["loan_limit"].quantile(0.25)

IQR = Q3 - Q1

# 각 값의 이상치 여부 판별 (apply)

def is_outlier(value) :
    if (value > Q3 + 1.5*IQR) | (value < Q1 - 1.5*IQR) :
        return True
    else : 
        return False
        
# apply 함수를 통하여 각 값의 이상치 여부를 찾고 새로운 열에 결과 저장
train_dset["loan_limit_outlier"] = train_dset["loan_limit"].apply(lambda x : is_outlier(x)) # axis = 1 지정 필수

loan_limit_not_outlier = train_dset["loan_limit"][train_dset["loan_limit_outlier"] == False]

sns.distplot(loan_limit_not_outlier, kde = True, rug = True)
```

## Label Encoding

```python
# label encoding
class MultiColumnLabelEncoder:
    def __init__(self,columns = None):
        self.columns = columns # array of column names to encode

    def fit(self,X,y=None):
        return self # not relevant here

    def transform(self,X):
        '''
        Transforms columns of X specified in self.columns using
        LabelEncoder(). If no columns specified, transforms all
        columns in X.
        '''
        output = X.copy()
        if self.columns is not None:
            for col in self.columns:
                output[col] = LabelEncoder().fit_transform(output[col])
        else:
            for colname,col in output.iteritems():
                output[colname] = LabelEncoder().fit_transform(col)
        return output

    def fit_transform(self,X,y=None):
        return self.fit(X,y).transform(X)
```

```python
# label encoding valuable names
label_encoding_list = ["income_type", "employment_type", "purpose", "houseown_type"]

train_dset_labelencoded = MultiColumnLabelEncoder(columns = label_encoding_list).fit_transform(train_dset)
```

## 결측치 대체 process 함수

```python
# mice 실습 
import miceforest as mf
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np

# Load data and introduce missing values
iris = pd.concat(load_iris(as_frame=True,return_X_y=True),axis=1)
iris['target'] = iris['target'].astype('category')
iris_amp = mf.ampute_data(iris,perc=0.25,random_state=1991)
```

```python
# 1. Regression
# 2. Multiple Imputation by Chained Equations 
# 3. RandomForest
def missing_value_processing(df, method) :
    # drop columns
    drop_columns = ["application_id", "loanapply_insert_time", "bank_id", "product_id", "user_id", "birth_year",
                "personal_rehabilitation_complete_yn", "date_cd", "time", "insert_time", "company_enter_month",
                   "age"]
    df = df.drop(labels = drop_columns, axis = 1)
    
    # drop rows ( < 1% )
    df.dropna(subset=["loan_limit", "loan_rate", "gender", "age_group"])
    
    # 0 imputation
    df["personal_rehabilitation_yn"].fillna(0, inplace = True)
    
    df_column_list = list(df.columns)
    
    # imputation
    if method == "Reg" :
        linearrg = LinearRegression()
        imp = IterativeImputer(estimator = linearrg, missing_values = np.nan, max_iter = 5, verbose = 2, 
                              imputation_order = "roman", random_state = 72220219)
        ImputedData = imp.fit_transform(df)
        
        return pd.DataFrame(ImputedData, columns = df_column_list)
    
    elif method == "mice" :
        
        train_dset_labelencoded["income_type"] = train_dset_labelencoded["income_type"].astype(str)
        train_dset_labelencoded["employment_type"] = train_dset_labelencoded["employment_type"].astype(str)
        train_dset_labelencoded["personal_rehabilitation_yn"] = train_dset_labelencoded["personal_rehabilitation_yn"].astype(str)
        train_dset_labelencoded["purpose"] = train_dset_labelencoded["purpose"].astype(str)
        train_dset_labelencoded["gender"] = train_dset_labelencoded["gender"].astype(str)
        train_dset_labelencoded["age_group"] = train_dset_labelencoded["age_group"].astype(str)
        train_dset_labelencoded["houseown_type"] = train_dset_labelencoded["houseown_type"].astype(str)
        
        kds = mf.ImputationKernel(df,
                                 datasets = 5,
                                 save_all_iterations = True,
                                  categorical_feature = ["gender","is_applied", "income_type",
                                                         "employment_type", "houseown_type", "purpose",
                                                        "personal_rehabilitation_yn", "age_group"],
                                  random_state = 72220219
                                 )
        
        # Run the MICE algorithm for 6 iterations
        kds.mice(6, verbose = True)
        
        print(kds)
        
        # Return the completed dataset
        Final_results = pd.concat([kds.complete_data(i) for i in range(5)]).groupby(level = 0).mean()
        imputed_values_distribution = kds.impute_new_data(df)
        
        imputed_values_distribution.plot_imputed_distributions(wspace=0.35,hspace=0.4)
        
        return Final_results
    
    
    
```
